```python
import keras
from keras.models import load_model
```

```python
model=load_model("D:\\PROJECTS FINAL YEAR\\vgg16.h5")
```

```python
import cv2 
import numpy as np
```

```python
img = cv2.imread('D:\PROJECTS FINAL YEAR\Training\pituitary_tumor\p (33).jpg')
img = cv2.resize(img,(224,224))
img_array = np.array(img)
img_array.shape
```

```python
img_array = img_array.reshape(1,224,224,3)
img_array.shape
```

```python
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img('D:\PROJECTS FINAL YEAR\Training\pituitary_tumor\p (33).jpg')
plt.imshow(img,interpolation='nearest')
plt.show()
```

```python
a=model.predict(img_array)
indices = a.argmax()
indices
```

```python
if(indices==0 ):
    print("Glimola")
elif(indices==1):
    print("Meningioma")
elif(indices==2):
    print("No tumor")
elif(indices==3):
    print("pituitarty")
else:
    print("Model not trained to identify this image")
    
    
```
